#Train set

```python
import pandas as pd
df_PIMA = pd.read_csv("/content/drive/MyDrive/Machine_learning/PIMA/PIMA_training.csv",)
df_PIMA.head()
```

```python
x_train=df_PIMA[[str(i) for i in range(8)]]
x_train.head()
```

```python
y_train=df_PIMA[["8"]]
y_train.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier as tree
```

```python
DT= tree()
DT.fit(x_train,y_train)
```

```python
DT.get_params()
```

#Test set

```python
df_Test = pd.read_csv("/content/drive/MyDrive/Machine_learning/PIMA/PIMA_testing.csv")
```

```python
df_Test.head()
```

```python
x_test = df_Test[[str(i) for i in range(8)]]
x_test.head()
```

```python
y_test = df_Test["8"]
y_test.head()
```

```python
features_names=x_train.columns
```

```python
from sklearn import tree as tr
from matplotlib import pyplot as plt 

fig =plt.figure(figsize=(25,20))

_=tr.plot_tree(DT,
              feature_names=features_names,
              class_names={0.0:"0",1.0:"1"},
              filled=True,
              fontsize=12)
              

plt.show()
```

#Predicción y probabilidad del modelo
Debido a que no hemos selecionado criterios de para del árbol, este ha crecido automáticamente hasta donde es capas de distinguir cada clase con seguridad en función del train-set, esto se puede ver en que dara resultados con probabilidad únicamente de 0 y 1.


Esto puede traer problemas de sobre ajuste y que al tratar de predecir un sobre test de valores errados.

Además aunque el modelo tenga buenos resultados en test, puede que sea tan grande que pierda interpretabilidad, si esto es un criterio importanto de nuestro modelo, puede que nos interese podarlo o fijar a cierto valor la profundidad.

```python
DT.predict_proba(x_test)[0:7]
```

```python
test_score=DT.score(x_test, y_test)
train_score=DT.score(x_train, y_train)
print(f"Score con train : {train_score}")
print(f"Score con test : {test_score}")
```

##Evaluación 
Aquí se puede ver como el modelo ha sido sobre etrenado, se ajusta demaciado bien a los datos de train y muy mal a los de test.

Acontinuación trataremos de ajustar la profundidad del árbol para obtener mejores resultados.

```python
DT2= tree(max_depth=6)
DT2.fit(x_train,y_train)
```

```python
DT2.get_params()
```

```python
DT2.predict_proba(x_test)[0:7]
```

#Predicción y probabilidad del modelo con menos profundidad
Se puede ver como ahora al tener menos profundiad el modelo, no está completamente seguro de qué etiqueta le corresponde a cada dato, ha cierta incertidumbre.

Si la profundidad es la correcta el árbol dara buenos resultados en test, pero existe la posibilidad de que no tenga la suficiente y no pueda predecir correctamente.

Tener un árbol con una profundiad dentro de un rango considerable puede ayudar si interesa la interpretabilidad del modelo para alguna aplicación, pero se tendra que tener cuidado a no sub-entrenar el modelo.

#Evaluación del árbol con menos profundiad
Se puede ver como se ha empeorado el rendimiento en train, pero se ha beneficiado en test.

```python
test_score2=DT2.score(x_test, y_test)
train_score2=DT2.score(x_train, y_train)
print(f"Score con train : {train_score2}")
print(f"Score con test : {test_score2}")
```

```python
fig =plt.figure(figsize=(25,20))

_=tr.plot_tree(DT2,
              feature_names=features_names,
              class_names={0.0:"0",1.0:"1"},
              filled=True,
              fontsize=12)
              
#tr.plot_tree(DT5)
plt.show()
```

#Sub-entrenamiento
Apesar de que interese la interpretabilidad del árbol y busquemos simplificarlo para lograr esto, se puede llegar al extremo de sub-entrenarlo.

En estos casos si el árbol no tiene la suficiente profundidad, no sera capaz de predecir correctamente tanto para train como para test


```python
DT3= tree(max_depth=1)
DT3.fit(x_train,y_train)
```

```python
DT3.get_params()
```

```python
DT3.predict_proba(x_test)[0:7]
```

```python
test_score3=DT3.score(x_test, y_test)
train_score3=DT3.score(x_train, y_train)
print(f"Score con train : {train_score3}")
print(f"Score con test : {test_score3}")
```

```python
fig =plt.figure(figsize=(25,20))

_=tr.plot_tree(DT3,
              feature_names=features_names,
              class_names={0.0:"0",1.0:"1"},
              filled=True,
              fontsize=12)
              
#tr.plot_tree(DT5)
plt.show()
```

#Optimización de parámetros 
Usaremos GridSearch para buscar los parámetros que nos den un mejor desempeño del modelo, para evaluar estos parámetros se utiliza validación cruzada (CV).

```python
from sklearn.model_selection import GridSearchCV
```

```python
DT4=tree()
parameters = {'max_depth': [i for i in range(2,31)] ,'criterion':['gini', 'entropy']}
DT_grid = GridSearchCV(DT4,parameters, cv=5,)
#              refit=True, verbose=0)
DT_grid.fit(x_train, y_train)
```

```python
print(DT_grid.best_params_)
print(DT_grid.best_estimator_)
```

```python
test_scoreG=DT_grid.score(x_test, y_test)
train_scoreG=DT_grid.score(x_train, y_train)
print(f"Score con train : {train_scoreG}")
print(f"Score con test : {test_scoreG}")
```

```python
param=DT_grid.best_params_
```

```python
DT5=tree(criterion=param["criterion"],max_depth=param["max_depth"])
DT5.fit(x_train, y_train)
```

```python
test_score5=DT5.score(x_test, y_test)
train_score5=DT5.score(x_train, y_train)
print(f"Score con train : {train_score5}")
print(f"Score con test : {test_score5}")
```

```python

fig =plt.figure(figsize=(25,20))

_=tr.plot_tree(DT5,
              feature_names=features_names,
              class_names={0.0:"0",1.0:"1"},
              filled=True,
              fontsize=12)
              
#tr.plot_tree(DT5)
plt.show()
```
